```
data = requests.get("https://api.covid19india.org/raw_data.json").json()
data = pd.DataFrame.from_dict(data['raw_data'])
data.to_csv("covid.csv")
```

```
df_covid = pd.read_csv("covid.csv")
```

```
#dropping unnamed column as it is not providing us any information
df_covid.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```
df_covid.isnull().sum()
```

```
#Finding the percentage of missing values
(df_covid.isnull().sum() * 100 / df_covid.index.size).round(2)
```

```
df_covid.duplicated().sum()
```

```
df_covid.duplicated()
```

```
#Using missing category imputation method to fill missing values in gender column as missing
df_covid["gender"].fillna("Missing", inplace = True) 
df_covid 
```

```
sns.catplot(x="gender", y="patientnumber", data=df_covid, kind="box")
```

```
#filling missing values in "backupnotes" column to "Travel hsitory unknown" because we donot have enough information on this 
#and the column has 27822 NaN vallues i.e 98.71% of the data is missing.
df_covid["backupnotes"].fillna("Travel history unknown", inplace = True) 
df_covid 
```

```
df_covid.isnull().sum()
```

```
df_covid["contractedfromwhichpatientsuspected"].mode()
```

```
#Filling the missing nationality using mode because this covid anlysis dataset has records of India only, so majority of the affected people belong to india
data = df_covid["nationality"].mode()[0]
df_covid["nationality"].fillna(data, inplace=True)
df_covid
```

```
#Using missing category imputation method to fill missing values in contractedfromwhichpatientsuspected column as missing
df_covid["contractedfromwhichpatientsuspected"].fillna("Missing", inplace = True) 
```

```
#Since there was only 1 missing value in "detectedstate"
missing_state = df_covid["detectedstate"].value_counts()[0]
df_covid["detectedstate"].fillna(missing_state, inplace=True)
```

```
df_covid.groupby("gender")["detectedstate"].value_counts().groupby(level=0).head(3)
```

```
#Since we don't know the detectedcity or detectedstate and replacing it with any other method felt irrelevant
df_covid["detectedcity"].fillna("unknown", inplace = True) 
df_covid["detecteddistrict"].fillna("unknown", inplace = True) 
```

```
#Drooping estimatedonsetdate column since it has 100% missing values
df_covid.drop(["estimatedonsetdate"], axis=1, inplace=True)
```

```
#Using arbitrary values imputation method to replace missing values, since patientnumber is unique and the distribution is positive using arbitrary value -1
df_covid["patientnumber"].fillna("-1", inplace = True)
```

```
missing_code = df_covid["statecode"].mode()[0]
df_covid["statecode"].fillna(missing_code, inplace = True)
```

```
#Since we don't know from which source we got the information
df_covid["source1"].fillna("Missing source", inplace = True)
df_covid["source2"].fillna("Missing source", inplace = True)
df_covid["source3"].fillna("Missing source", inplace = True)
```

```
#No notes on travel history, so replacing the missing values in notes column as Missing
df_covid["notes"].fillna("Missing", inplace = True)
```
